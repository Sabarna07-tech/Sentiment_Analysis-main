```python
docs=[]
for text,label in zip(df['Text'][number:],df['label'][number:]):
    doc = nlp(text)
    if label == 'Depression':
        doc.cats['Depression']=1
        doc.cats['Anxiety Disorder']=0
        doc.cats['Panic Disorder']=0
        doc.cats['Anger/ Intermittent Explosive Disorder']=0
        doc.cats['Narcissistic Disorder']=0
    if label == 'Anxiety Disorder':
        doc.cats['Depression']=0
        doc.cats['Anxiety Disorder']=1
        doc.cats['Panic Disorder']=0
        doc.cats['Anger/ Intermittent Explosive Disorder']=0
        doc.cats['Narcissistic Disorder']=0
    if label == 'Panic Disorder':
        doc.cats['Depression']=0
        doc.cats['Anxiety Disorder']=0
        doc.cats['Panic Disorder']=1
        doc.cats['Anger/ Intermittent Explosive Disorder']=0
        doc.cats['Narcissistic Disorder']=0
    if label == 'Anger/ Intermittent Explosive Disorder':
        doc.cats['Depression']=0
        doc.cats['Anxiety Disorder']=0
        doc.cats['Panic Disorder']=0
        doc.cats['Anger/ Intermittent Explosive Disorder']=1
        doc.cats['Narcissistic Disorder']=0
    if label == 'Narcissistic Disorder':
        doc.cats['Depression']=0
        doc.cats['Anxiety Disorder']=0
        doc.cats['Panic Disorder']=0
        doc.cats['Anger/ Intermittent Explosive Disorder']=0
        doc.cats['Narcissistic Disorder']=1
    docs.append(doc)
    
doc_bin = DocBin(docs=docs)
doc_bin.to_disk(r"./validation.spacy")
```

```python
text = df['Text'][781]
label = df['label'][781]
print(text)
print(label)
```

```python
def ANS(text:str)->str:
    doc=model(text)
    dic = doc.cats
    ans = max(zip(dic.values(),dic.keys()))[1]
    return ans
```
