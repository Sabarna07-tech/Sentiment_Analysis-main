```python
import re
import pandas as pd
import contractions as con
import numpy as np
```

```python
link = "https:\/\/[^ ]{1,}"
contraction = "[A-Za-z]*'[A-Za-z]{1,2}"
punctuation = list("""!"#$%&'()*+,-./:;<=>?@[\]^_`{|}~""")
```

```python
df = pd.read_csv('train.csv')
df
```

```python
df.drop(['Unnamed: 0'],axis=1,inplace=True)
df
```

```python
for i in range(len(df)):
    df.loc[i,'Text']=df.loc[i,'Text'].lower()
df
```

```python
for i in range(len(df)):
    t = df.loc[i,'Text']
    L = re.findall(string=t,pattern=contraction)
    if len(L)>0:
        for word in L:
            df.loc[i,'Text'] = re.sub(re.escape(word),con.fix(word), t)
df
```

```python
class DecodeContraction:
    def __init__(self):
        import re
        import contractions
        self.__contraction_dict={'ve':' have','s':' is','d':' had','r':' are'}
        self.__pattern = "[a-z]*'[a-z]{1,2}"
        
    def fit(self,text:str):
        self.__list = re.findall(string=text,pattern=self.__pattern)
    
    def convert(self,text:str):
        for word in self.__list:
            main,key = word.split("'")
            try:
                new_word = main + self.__contraction_dict[key]
                text = re.sub(re.escape(word),new_word,text)
            except:
                pass
        self.NEWSTRING_ = text
```

```python
for i in range(len(df)):
    t = df.loc[i,'Text']
    convertor = DecodeContraction()
    convertor.fit(t)
    convertor.convert(t)
    df.loc[i,'Text'] = convertor.NEWSTRING_
df
```

```python
for i in range(len(df)):
    t = df.loc[i,'Text']
    L = re.findall(string=t,pattern=link)
    if len(L)>0:
        for word in L:
            df.loc[i,'Text'] = re.sub(re.escape(word),'', t)
df
```

# LsTM


```python
from tensorflow.keras.layers import Embedding,LSTM,Dense,Dropout
from tensorflow.keras.models import Sequential
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.preprocessing.text import one_hot
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.regularizers import L2
from tensorflow.keras.constraints import Constraint,MaxNorm
```

```python
voc_size = 6000
```

```python
dic = {'Depression':0,'Anxiety Disorder':1,'Panic Disorder':2,'Anger/ Intermittent Explosive Disorder':3,'Narcissistic Disorder':4}
```

```python
df['label'] = df['label'].map(dic)
df
```

```python
y=[]
for each in df['label']:
    l=[0,0,0,0,0]
    l[each]=1
    y.append(l)
y=np.array(y)
y
```

# one hot representation

```python
x=df['Text'].copy()
x
```

```python
from nltk.stem import PorterStemmer
```

```python
stemmer = PorterStemmer()
```

```python
for i in range(len(x)):
    t=x[i]
    l=''
    for w in t.split():
        l+=' '+stemmer.stem(w)
    x[i]=l
x
```

```python
onehot_rep = [one_hot(words,voc_size) for words in x]
onehot_rep
```

```python
max_len = 0
for each in onehot_rep:
    max_len = max(max_len,len(each))
max_len
```

```python
padded_sequence = pad_sequences(onehot_rep,padding='post',maxlen=max_len)
padded_sequence
```

```python
max_word=0
for each in padded_sequence:
    for w in each:
        max_word = max(max_word,w)
max_word
```

```python
padded_sequence = padded_sequence/max_word
padded_sequence
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(padded_sequence,y,random_state=42,test_size=0.2)
```

```python
print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

# Model:

```python
model = Sequential()
model.add(Embedding(voc_size,40,input_length=max_len))
model.add(LSTM(300))
model.add(Dense(20,activation='relu',kernel_regularizer=L2(1e-6),kernel_constraint=MaxNorm(3)))
model.add(Dropout(0.7))
model.add(Dense(20,activation='relu',kernel_regularizer=L2(1e-6),kernel_constraint=MaxNorm(3)))
#model.add(Dropout(0.7))
model.add(Dense(5,activation='softmax',kernel_constraint=MaxNorm(3),kernel_regularizer=L2(1e-5)))
```

```python
model.compile(loss='categorical_crossentropy',optimizer=Adam(learning_rate=0.01),metrics=['accuracy'])
```

```python
print(model.summary())
```

```python
epochs=0
```

```python
epoch=10
model.fit(x_train,y_train,validation_data=(x_test,y_test),epochs=epoch,batch_size=64)
epochs+=epoch
```
